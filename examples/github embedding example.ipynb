```python
from graphembedding.playground import load_github
from graphembedding import complEx
```

```python
# Load Sample dataset 
github_dataset = load_github() 
triplets = github_dataset[['subject','relation','object']].values
```

```python
# That's all. One line code.
# It takes about 15 minutes per epoch.
node_embedding, edge_embedding = complEx(triplets, 
                                         num_epochs=10, verbose=2)
```

```python
node_embedding.head()
```

```python
edge_embedding.head()
```

### find Similiar repository using Embedding

```python
repo_embedding = node_embedding[
    node_embedding.index.map(lambda x: isinstance(x, str))]
repo_embedding.head()
```

```python
target = repo_embedding.loc['tensorflow/tensorflow']
(
    repo_embedding
    .dot(target)
    .sort_values(ascending=False)
    .iloc[:10]
)
```
